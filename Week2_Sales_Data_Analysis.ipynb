```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)
records = 50

data = {
    "order_id": range(1001, 1001 + records),

    # Visible date format
    "order_date": pd.date_range(
        start="2025-01-01",
        periods=records,
        freq="D"
    ).strftime("%Y-%m-%d"),

    "product_name": np.random.choice(
        ["Laptop", "Mouse", "Keyboard", "Office Chair", "Notebook", "Printer"],
        records
    ),

    "category": np.random.choice(
        ["Electronics", "Furniture", "Stationery"],
        records
    ),

    "customer_name": np.random.choice(
        ["Amit", "Neha", "Rahul", "Sneha", "Kiran", "Priya"],
        records
    ),

    "region": np.random.choice(
        ["North", "South", "East", "West"],
        records
    ),

    "payment_mode": np.random.choice(
        ["Cash", "UPI", "Card", "NetBanking"],
        records
    ),

    "quantity": np.random.randint(1, 10, records),
    "unit_price": np.random.randint(100, 60000, records),

    # Discount in percentage
    "discount_percent": np.random.choice([0, 5, 10, 15], records)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Total before discount
df["total_amount"] = df["quantity"] * df["unit_price"]

# Discount value
df["discount_value"] = (df["discount_percent"] / 100) * df["total_amount"]

# Final amount after discount
df["final_amount"] = df["total_amount"] - df["discount_value"]

df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df.to_csv("week2_sales_analysis_output.csv", index=False)
```
